```python
import numpy as np
import cv2
#from matplotlib import pyplot as plt
import random 
import math

img=cv2.imread('baby512.jpg',0)

def readGrayscaleImage(img):
    GrayImage=img.copy()
    cv2.imwrite('GrayScaleImage.jpg',GrayImage)
    cv2.imshow('GrayScaleImage',GrayImage)
    cv2.waitKey(0)
    return GrayImage


def corruptImageSP(GrayImage):
    #Adding salt and pepper noise to image
    #img=cv2.imread('baby.jpg',0) 
    copyImg=GrayImage.copy()
    NoisyImage=np.zeros(copyImg.shape,np.uint8)
    for i in range(copyImg.shape[0]):  #traverse rows of matrix
        for j in range(copyImg.shape[1]):  #traverse columns of matrix
            a=random.random()
            if a<0.90:
                NoisyImage[i][j]=copyImg[i][j]
            elif (a<0.98 and a>0.95): 
                NoisyImage[i][j]=0
            else: 
                NoisyImage[i][j]=255
    cv2.imwrite('Salt&Pepper Noisy image.jpg',NoisyImage)
    cv2.imshow('Salt&Pepper Noisy image',NoisyImage)
    cv2.waitKey(0)
    return NoisyImage

def noisereductionAveraging(GrayImage):
    #Noise reduction-smoothing by averaging
    copyImg=GrayImage.copy()
    SmoothImageAveraging=copyImg.copy()
    for i in range(1,copyImg.shape[0]-1):
        for j in range(1,copyImg.shape[1]-1):
            sum=0
            sum=(copyImg[i][j]//9)+(copyImg[i][j+1]//9)+(copyImg[i][j-1]//9)+(copyImg[i-1][j]//9)+(copyImg[i-1][j+1]//9)+(copyImg[i-1][j-1]//9)+(copyImg[i+1][j]//9)+(copyImg[i+1][j+1]//9)+(copyImg[i+1][j-1]//9)             
            SmoothImageAveraging[i][j]=sum
    cv2.imshow('SmoothImageAveraging',SmoothImageAveraging)
    cv2.imwrite('SmoothImageAveraging.jpg',SmoothImageAveraging)
    cv2.waitKey(0)
    return SmoothImageAveraging


def noisereductionMedianFiltering(GrayImage):
    #Noise reduction-smoothing by median filtering
    copyImg=GrayImage.copy()
    SmoothImageMedianFilter=copyImg
    for i in range(1,copyImg.shape[0]-1):  #traverse rows of matrix
        for j in range(1,copyImg.shape[1]-1):  #traverse columns of matrix
            kernel=[]
            kernel=[copyImg[i][j],copyImg[i-1][j],copyImg[i+1][j],copyImg[i][j-1],copyImg[i][j+1],copyImg[i-1][j-1],copyImg[i+1][j-1],copyImg[i-1][j+1],copyImg[i+1][j+1]]
            kernel.sort()
            SmoothImageMedianFilter[i][j]=kernel[4]              
    cv2.imshow('SmoothImageMedianFilter',SmoothImageMedianFilter)
    cv2.imwrite('SmoothImageMedianFilter.jpg',SmoothImageMedianFilter)
    cv2.waitKey(0)
    return SmoothImageMedianFilter

def binaryImageSimpleThresholding(GrayImage):
    #Grayscale to binary image by using simple thresholding
    binImgSimpThreshldngimg=GrayImage.copy() 
    for i in range(binImgSimpThreshldngimg.shape[0]):
        for j in range(binImgSimpThreshldngimg.shape[1]):
            if np.any(binImgSimpThreshldngimg[i][j]>80): 
                binImgSimpThreshldngimg[i][j]=255
            else: 
                binImgSimpThreshldngimg[i][j]=0

    cv2.imshow('Binary Image Simple Thresholding',binImgSimpThreshldngimg)
    cv2.imwrite('Binary Image Simple Thresholding.jpg',binImgSimpThreshldngimg)
    cv2.waitKey(0)
#     cv2.destroyAllWindows() 
    return binImgSimpThreshldngimg


def binaryImageP_tileThresholding(GrayImage):
    #Grayscale image to binary image by using P-Tile Thresholding
    binImgPTile=GrayImage.copy()
    width,depth=img.shape
    percentage=50/100
    thresholdpixelcount=width*depth*percentage  #pixel calculation from considered percentage value (this will correspond to threshold)
    #print(thresholdpixelcount)

    #Create hist datastructure
    hist=[]
    hist=np.zeros(256)
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            hist[img[i,j]]=hist[img[i,j]]+1
    #print(hist)        

    #Find intensity T correspondng to the percentage considered
    sum=0
    threshold=0
    for k in range(254,-1,-1):
        sum=sum+hist[k]
        if(sum>=thresholdpixelcount):
            break
    threshold=k

    #Convert into binary image
    for m in range(img.shape[0]):
        for n in range(img.shape[1]):
            if np.any(img[m][n]>threshold): 
                binImgPTile[m][n]=255
            else: 
                binImgPTile[m][n]=0

    cv2.imshow('Binary Image P-tile thresholding',binImgPTile)
    cv2.imwrite('Binary Image P-tile thresholding.jpg',binImgPTile)
    cv2.waitKey(0)
#     cv2.destroyAllWindows() 
    return binImgPTile

def binaryImageIterativeThresholding(GrayImage):
    #Grayscale image to binary using Iterative thresholding
    binImgItrThrs=GrayImage.copy()
    threshold=110
    revisedthreshold=0
    tolerance=10
    region1 = []
    region2 = []
    while(abs(revisedthreshold - threshold) > tolerance):
        for i in range(GrayImage.shape[0]):
            for j in range(GrayImage.shape[1]):
                if(GrayImage[i][j]>threshold):
                    region1.append(GrayImage[i][j])
                else:
                    region2.append(GrayImage[i][j])

        avg_region1 = sum(region1)//len(region1)
        avg_region2= sum(region2)//len(region2)
        revisedthreshold = (avg_region1+avg_region2)//2
        if(abs(revisedthreshold - threshold) < tolerance):
            break
        else:
            threshold=revisedthreshold

    for i in range(GrayImage.shape[0]):
        for j in range(GrayImage.shape[1]):
            if(GrayImage[i][j]>threshold):
                binImgItrThrs[i][j]=255
            else:
                binImgItrThrs[i][j]=0

    cv2.imshow('Binary Image iterative thresholding',binImgItrThrs)
    cv2.imwrite('Binary Image iterative thresholding.jpg',binImgItrThrs)
    cv2.waitKey(0)
#     cv2.destroyAllWindows()
    return binImgItrThrs


def histogramGrayscale():
    #Histogram of Grayscale image
    img = cv2.imread('baby.jpg',0)
    plt.hist(img.ravel(),256,[0,256]); plt.show()


def histogramRGB():
    #Histogram of RGB image
    img = cv2.imread('baby.jpg')
    color = ('b','g','r')
    for i,col in enumerate(color):
        histr = cv2.calcHist([img],[i],None,[256],[0,256])
        plt.plot(histr,color = col)
        plt.xlim([0,256])
    plt.show()



def main():
    GrayImage=readGrayscaleImage(img)
    NoisyImage=corruptImageSP(GrayImage)
    SmoothImageAveraging=noisereductionAveraging(NoisyImage)
    SmoothImageMedianFilter=noisereductionMedianFiltering(NoisyImage)
    binImgSimpThreshldngimg=binaryImageSimpleThresholding(GrayImage)
    binImgPTile=binaryImageP_tileThresholding(GrayImage)
    binImgItrThrs=binaryImageIterativeThresholding(GrayImage)
    histogramGrayscale()
    histogramRGB()
  
    

main()
```
